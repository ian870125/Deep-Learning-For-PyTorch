```python
import torch
import torchvision
import matplotlib.pyplot as plt
```

```python
use_gpu = True if torch.cuda.is_available() else False
model = torch.hub.load('facebookresearch/pytorch_GAN_zoo:hub', 'PGAN', model_name = 'celebAHQ-512', pretrained = True, useGPU = use_gpu)
```

```python
num_images = 4
noise, _ = model.buildNoiseData(num_images)
with torch.no_grad():
    generated_images = model.test(noise)
    grid = torchvision.utils.make_grid(generated_images.clamp(min=-1, max=1)
                                   , scale_each=True, normalize=True)
plt.imshow(grid.permute(1, 2, 0).cpu().numpy())
plt.axis('off');
```

```python
class Solution:
    def twoSum(self, nums, target: int):
        return [[nums.index(i), nums.index(target - i)] for i in nums if (target - i) in nums and (target - i) != i][0]
```
